```python
import math
import pandas as pd

# create data
d = {
    'KW': ['hard drive', 'ssd hard drive', 'laptop hd', 'hd', 'ssd'],
    'SFR': [1, 10, 100, 100, 1000],
    'CS': [0.05, 0.05, 0.05, 0.05, 0.05]
}

# generate dataframe
df = pd.DataFrame(data=d)

# output dataframe
df
```

```python
# loop over dataframe and run exponential decay on each row values
for index, row in df.iterrows():
    print("{:.5f}".format(math.pow(row["SFR"], row["CS"])))
```
